## Panda DataFrames Part-3
### Multi-Index and Index Hierarchy

```python
#index Levels
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(hier_index)
```

```python
list(zip(outside,inside))
```

```python
hier_index
```

```python
df = pd.DataFrame(randn(6,3),hier_index,['A','B','C'])
randn(6,3) #Random number form normal distributions of 6 rows and 3 coloums
```

```python
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```
