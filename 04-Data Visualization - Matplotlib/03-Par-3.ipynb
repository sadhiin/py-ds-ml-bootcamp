```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=1,alpha=0.8,linestyle=':',marker='+',markersize=10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green',lw=2,ls='--')

#x or y axis limit setting
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
