## Functional Method 
Not a good way to do the work. But we can use methods on the go. 

```python

pl.plot(x,y)
```

```python
pl.plot(x,y,'r') # r is the color red
pl.xlabel('X Label')
pl.ylabel('Y Label')
pl.title('This is title')
```

```python
pl.subplot(1,2,1)
pl.plot(x,y,'r')

pl.subplot(1,2,2)
pl.plot(y,x,'b')

# pl.subplot(2,1,1)
# pl.plot(y,x/y,'y')
```

## Object Oriented Method
Better way to use Matplotlib

```python
fig = pl.figure()
#creating the figure object 

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #([left,bottom,width,hight])
#fillwing the canvace with given values

axes.plot(x,y)
#ploting the values

axes.set_xlabel('X-Lable')
axes.set_ylabel('Y-Lable')
axes.set_title('Title')
```

```python
fig = pl.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x,y)
axes2.plot(y,x,'r')

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```
