```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y= x**2
```

```python
# fig, axes = plt.subplots()
# fig,axes=plt.subplots(nrows=3,ncols=3)
fig,axes=plt.subplots(nrows=1,ncols=2)
# axes.plot(x, y)
# for current_ax in axes:
#     current_ax.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title("First Plot")
axes[1].plot(y,x)
axes[1].set_title("Second Plot")


plt.tight_layout()    
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(5,3))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,4))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig=plt.figure()
ax =fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X squer')
ax.plot(x,x**3,label='X cube')

ax.set_title('Title')
# ax.set_xlabel('X Label')
# ax.set_ylabel('Y Lable')

ax.legend() #[loc= 0 to 10 value specifi the location of the labels of line]
ax.legend(loc=10)
```
